```python

import cv2                 
import numpy as np         
import os                  
from random import shuffle 
from tqdm import tqdm      
import matplotlib
import cv2    
from sklearn import decomposition    
import matplotlib.pyplot as plt 
   
```

```python

img = cv2.imread('D://All/00000417_002.png',0)  
vmax = max(img.max(), -img.min())


n_components = 20

estimator = decomposition.NMF(n_components = n_components, init = 'random', tol=5e-3)    
W = estimator.fit_transform(img)
H = estimator.components_

new_img = np.dot(W,H)
cv2.imwrite('E://1.png',new_img)
```

```python
new_img = cv2.imread('E:/1.png',0)
new_img2=cv2.imread('E:/1.png')

lab= cv2.cvtColor(new_img2, cv2.COLOR_BGR2LAB)


l, a, b = cv2.split(lab)


clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl = clahe.apply(l)

limg = cv2.merge((cl,a,b))

final = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)
```

```python
plt.imshow(final,cmap='gray')
```

```python
equ = cv2.equalizeHist(new_img)

cv2.imwrite('E://Original.png',img)
cv2.imwrite('E://nmf.png',new_img)
cv2.imwrite('E://equ.png',equ)
cv2.imwrite('E://enhanced.png',final)
#plt.imshow(equ,cmap='gray')

enhanced=cv2.imread('E:/enhanced.png',0)

res = np.hstack((img,new_img,equ,enhanced)) #stacking images side-by-side
```
